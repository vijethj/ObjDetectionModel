```python
import os
import random
import shutil
```

```python
csv_folder_path = "F:\Sem6\TDL\hackathon1\Training Dataset_csv" #original dataset (put cleaned if you've cleaned it)
mp4_folder_path = "F:\Sem6\TDL\hackathon1\Training Dataset_Videos" #mp4
jpg_folder_path = "F:\Sem6\TDL\hackathon1\Train_data" #jpg 
new_folder_path = "F:\Sem6\TDL\hackathon1\Retained data"
```

```python
csv_files = [f for f in os.listdir(csv_folder_path) if f.endswith('.csv')]
```

```python
random.shuffle(csv_files)
selected_csv_files = csv_files[:int(len(csv_files) * 0.3)]
print(f"Selected CSV files: {selected_csv_files}")
```

```python
for csv_file in selected_csv_files:
    csv_file_path = os.path.join(csv_folder_path, csv_file)
    identifier = csv_file.split('_LABELS')[0]
    mp4_file = f"{identifier}.mp4"
    mp4_file_name, mp4_file_ext = os.path.splitext(mp4_file)
    new_mp4_file_name = mp4_file_name + '.mp4'
    print(f"Selected mp4 files: {new_mp4_file_name}")
    mp4_file_path = os.path.join(mp4_folder_path, new_mp4_file_name)
    for i in range(1, 401):
        jpg_file = "{}_{}.jpg".format(identifier, i)
        if os.path.exists(os.path.join(jpg_folder_path, jpg_file)):
            print("Match found: {} and {}".format(csv_file, jpg_file))
    #jpg_file = csv_file.replace("_LABELS.csv", identifier2)
    #jpg_file_name, jpg_file_ext = os.path.splitext(jpg_file)
    #new_jpg_file_name = jpg_file_name + '.jpg'
    print(f"Selected jpg files: {jpg_file}")
    jpg_file_path = os.path.join(jpg_folder_path, jpg_file)
    print(f"Started")

    if os.path.exists(mp4_file_path) or os.path.exists(jpg_file_path):
        print(f"Started???")
        new_csv_path = os.path.join(new_folder_path, csv_file)
        new_mp4_path = os.path.join(new_folder_path, new_mp4_file_name)
        new_jpg_path = os.path.join(new_folder_path, jpg_file)
        shutil.copy2(csv_file_path, new_csv_path)
        shutil.copy2(mp4_file_path, new_mp4_path)
        print(f"Worked")
        shutil.copy2(jpg_file_path, new_jpg_path)
    else:
        print(f"Something went wrong")
```

```python
# Step 4: Delete the excess data
for f in os.listdir(csv_folder_path):
    if f.endswith('.csv') and f not in selected_csv_files:
        os.remove(os.path.join(csv_folder_path, f))
for f in os.listdir(mp4_folder_path):
    if f not in selected_csv_files:
        os.remove(os.path.join(mp4_folder_path, f))
for f in os.listdir(jpg_folder_path):
    if f not in selected_csv_files:
        os.remove(os.path.join(jpg_folder_path, f))
```
