```python
import os
import cv2
import pandas as pd
```

```python
input_dir = "F:\Sem6\TDL\hackathon1\Training Dataset_Videos"
output_dir = "F:\Sem6\TDL\hackathon1\Train_data"
```

```python
for input_file in os.listdir(input_dir):
    op_dir = output_dir
    ip_file = os.path.join(input_dir, input_file)

    print(ip_file)
    print(op_dir)
    cap = cv2.VideoCapture(ip_file)
    success, image = cap.read()
    count = 1
    while success:
#     # Save the frame as an image in the output directory
        cv2.imwrite(os.path.join(op_dir, f'{input_file[:-4]}_{count}.jpg'), image)
# 		# Read the next frame
        success, image = cap.read()
# 		# Increment the frame counter
        count += 1

# 	# Release the video capture object:
    cap.release()
```
